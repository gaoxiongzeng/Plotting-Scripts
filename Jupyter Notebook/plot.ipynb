```python
# Line Chart
%matplotlib inline

import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
import os

from matplotlib.patches import Ellipse

mpl.rcParams['font.family'] = 'Times New Roman'

# generate Figures Folder if not exists
if os.name == 'posix':
    output_folder = 'figures'
else:
    output_folder = "d:\\temp"

colors = ['grey', '#8899ff', 'cyan', 'orange', 'yellow', 'magenta', 'red']


DATA_X = np.array( [20, 40, 80, 160, 320] )

DATA_Y1 = np.array([313, 463, 665, 804, 899])

DATA_Y2 =   np.array([389, 693, 858, 932, 938])

    
def draw_all(title="Title"):
    fig, ax = plt.subplots(figsize = (5, 3.5))
    plt.plot( DATA_X, DATA_Y1, color='#303030', marker='o', linestyle='-', linewidth=2.0, markersize=9)
    plt.plot( DATA_X, DATA_Y2, color='#ff00ff', marker='s', linestyle='-', linewidth=2.0, markersize=9)
    ax.xaxis.grid(which='major',alpha=.6)
    ax.yaxis.grid(which='major',alpha=.6)

    # annotation
    ax.annotate('This is an annotation.',
                  xy=(DATA_X[0], DATA_Y2[0]), xycoords='data',
                  xytext=(50, 5), textcoords='offset points',
                  size=16,
                  bbox=dict(boxstyle="round",
                            fc=(1.0, 0.7, 0.7),
                            ec=(1., .5, .5)),
                  arrowprops=dict(arrowstyle="wedge,tail_width=1.", alpha = 0.6,
                                  fc=(1.0, 0.7, 0.7), ec=(1., .5, .5),
                                  patchA=None,
                                  patchB=Ellipse((2, -1), 0.5, 0.5),
                                  relpos=(0.2, 0.8),
                                  connectionstyle="arc3,rad=-0.1"))
    
    plt.legend(labels=['Scheme 1', 'Scheme 2'], 
           bbox_to_anchor=(1.0, 0.8), loc=1, ncol=1, fontsize=14)
    plt.ylabel('Y Label',
           fontsize=18,
           fontweight='bold')
    plt.xlabel('X Label',
           fontsize=18,
           fontweight='bold')
    plt.tick_params(axis='x',labelsize=18)
    plt.tick_params(axis='y',labelsize=18)
    plt.tight_layout()
    plt.savefig( output_folder + os.sep + title+'.pdf', format='pdf', bbox_inches='tight' ) # save as pdf
    plt.show()
    
draw_all("linechart")
```

```python
# Bar Plot
%matplotlib inline

import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
import os

# Hatch line width setting
mpl.rcParams['hatch.linewidth'] = 2.4
mpl.rcParams['font.family'] = 'Times New Roman'

# Number of Groups
n_groups = 1

# Figures Folder if not exists
if os.name == 'posix':
    output_folder = 'figures'
else:
    output_folder = "d:\\temp"


# data to draw
input_data = \
      [
          (49), 

          (54), 

          (55), 
          
          (53), 

          (56), 

          (53), 

          (35), 
      ]
        
def draw_one_group( data, title, its_xticks, data_size=7, legend_or_not=0):
    fig, ax = plt.subplots()
    
    index = np.arange(n_groups)
    bar_width = 0.05

    opacity = 0.4
    
    interval_between_bar = 0.2
    
    colors = ['#cccccc','grey', '#303030', '#00ffff', '#0080ff', '#0000ff', '#ff00ff']
    hatches= ['-', 'x', '//', '.', '\\', '+', '//']
    labels = ['Scheme A','Scheme B','Scheme C','Scheme D','Scheme E','Scheme F','Scheme G']
    
    for i in range(data_size):
        plt.bar(index + (1 + i + interval_between_bar*i)*bar_width, data[i], bar_width,
                 fill=True,
                 color=colors[i],
                 hatch=hatches[i],
                 label=labels[i],
                 edgecolor='black')

    plt.ylabel('Y Label',
           fontsize=27,
           fontweight='bold')
    plt.xlabel( its_xticks, fontsize=27, fontweight='bold')
    
    ax.yaxis.grid(alpha=.6)
    if legend_or_not:
        plt.legend(loc=9, fontsize=19, ncol=2)
        ax.set_ylim([0, int (2.5 * max([d for d in data]))])
    else:
        ax.set_ylim([0, int (1.2 * max([d for d in data]))])
    
    plt.tick_params(axis='y',labelsize=22)
    plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=False) # labels along the bottom edge are off
    plt.tight_layout()
    plt.savefig( output_folder + os.sep + title+'.pdf', format='pdf', bbox_inches='tight' ) # save as pdf
    plt.show()

draw_one_group (input_data, "barplot", "X Label", 7, 1)
```

```python
# Group Bar Plot
%matplotlib inline

import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
import os

# Hatch line width setting
mpl.rcParams['hatch.linewidth'] = 2.4
mpl.rcParams['font.family'] = 'Times New Roman'

# File Name
title = "group_barplot"

# Number of Groups
n_groups = 6

# Font weight bold
legend_properties = {'weight':'bold'}

# generate Figures Folder if not exists
if os.name == 'posix':
    output_folder = 'figures'
else:
    output_folder = "d:\\temp"


# data to draw
data_metric1 = [
          (16.6, 13.8, 12.3, 11.5, 11.2, 11),

          (16.6, 13.8, 12.3, 11.5, 11.1, 11),

          (16.5, 13.7, 12, 11.4, 11, 10.9),

      ]

data_metric2 = [
          (944, 936, 931, 896, 841, 824),

          (938, 926, 892, 849, 817, 794),

          (915, 912, 823, 761, 749, 712),

      ]


fig, ax = plt.subplots(figsize = (10, 7.5))
# subplot 1
plt.subplot(2, 1, 1)

index = np.arange(n_groups)
bar_width = 0.2
interval_between_bar = 0.1


SIZE = 3
colors = ['#00ffff', '#0080ff', '#0000ff']
hatches= ['-', 'x', '//', '.', '\\', '*']
labels = ['Scheme A', 'Scheme B', 'Scheme C']

for i in range(SIZE):
    plt.bar(index + (1 + i + interval_between_bar*i)*bar_width, data_metric1[i], bar_width,
                 fill=True,
                 #alpha=opacities[i],
                 color=colors[i],
                 hatch=hatches[i],
                 label=labels[i],
                 edgecolor='black')

plt.ylabel('Y Label 1',
           fontsize=26,
           fontweight='bold')

plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        labelbottom=False) # labels along the bottom edge are off
plt.grid(which='major',axis='y',alpha=.6)
plt.tick_params(axis='y',labelsize=22)
plt.legend(bbox_to_anchor=(0.23, 0.78), fontsize=20, ncol=3, 
        columnspacing=0.6, handletextpad=0.4)



# subplot 2
plt.subplot(2, 1, 2)

for i in range(SIZE):
    plt.bar(index + (1 + i + interval_between_bar*i)*bar_width,  data_metric2[i],  bar_width,
                 fill=True,
                 color=colors[i],
                 hatch=hatches[i],
                 label=labels[i],
                 edgecolor='black')


plt.ylabel('Y Label 2',
           fontsize=26,
           fontweight='bold')

plt.xlabel('X Label',
           fontsize=26,
           fontweight='bold')
plt.tick_params(
        axis='x',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        top=False,         # ticks along the top edge are off
    )
plt.xticks( index + (SIZE-0.5)*bar_width,
           ('1', '2', '4', '8', '16', '32'),
           fontsize=22)
plt.grid(which='major',axis='y',alpha=.6)
plt.tick_params(axis='y',labelsize=22)
plt.ylim([0, 1000])

plt.tight_layout()
plt.savefig( output_folder + os.sep + title+'.pdf', format='pdf', bbox_inches='tight' ) # save as pdf
plt.show()
```

```python
# CDF plot using hist()
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

x1 = np.loadtxt('./data/cdf_data1', delimiter=' ', unpack=True)
x2 = np.loadtxt('./data/cdf_data2', delimiter=' ', unpack=True)
n_bins = 100

fig, ax = plt.subplots(figsize=(8, 4))

# plot the cumulative histogram
n, bins, patches = ax.hist(x1, n_bins, density=True, histtype='step',
                           cumulative=True, label='Scheme A', linewidth=2)
n, bins, patches = ax.hist(x2, n_bins, density=True, histtype='step',
                           cumulative=True, label='Scheme B', linewidth=2)

# tidy up the figure
ax.grid(True)
ax.legend(loc='upper left')
#ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('CDF')

plt.savefig('./figures/cdf.pdf') 
plt.show()
```
